```python
# First separating the age groups

children = []
teens = []
adults = []
elderly = []

for e, i in enumerate(cleaned_age_data["Age"]):
    if i>0 and i<12:
        children.append(i)
    elif i>=12 and i<=17:
        teens.append(i)
    elif i>17 and i<=64:
        adults.append(i)
    elif i>64:
        elderly.append(i)
```

```python
# Converted all generated lists to numpy arrays

child = np.array(children)
teen = np.array(teens)
adult = np.array(adults)
elder = np.array(elderly)

# Calculating for the average ages for each set conveting the datatype to int as age has no float datatype
child_mean_age = int(np.mean(children))
teen_mean_age = int(np.mean(teens))
adult_mean_age = int(np.mean(adults))
elder_mean_age = int(np.mean(elderly))

age_mean_list = [child_mean_age,teen_mean_age,adult_mean_age,elder_mean_age]
age_ranges = ["(1-11)Years","(12-17)Years","(18-64)Years","(65+)Years"]
```

```python
# Since we have found out from publications that under the law 'Jedi' is not a religion, we would classify them as 'None' 
# While for ' ' we would classify them as 'Undecided'
# Agnostic is set to Undecided because they are persons with as particular mindset of being in between believing 
# and not believing in the existence of God. While private is set to undecided as I was unable to find any religion 
# called private in the UK. Persons below 18 years are not eligible to make decision on their religious preference.
# They were initially set to "NaN" but I had to set them to "Undecided" owing the fact that they are yet to make a decison
# about their religious preference and this might be done in the near future.

cleaned_age_data["Religion"].replace('Jedi','None', inplace=True)    
cleaned_age_data["Religion"].replace(' ','Undecided', inplace=True)
cleaned_age_data["Religion"].replace(np.nan,'Undecided', inplace=True)
cleaned_age_data["Religion"].replace('Agnostic','Undecided', inplace=True)
cleaned_age_data["Religion"].replace('Private','Undecided', inplace=True)
```
